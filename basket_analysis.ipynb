```python
import numpy as np 
import pandas as pd
```

```python
df = pd.read_csv('datasets\Groceries.csv', header=None)
```

```python
rows = len(df)
nos = np.arange(1,rows+1,1)
print(rows)
print(nos)
```

```python
df['id'] = nos
```

```python
temp = pd.melt(df,id_vars='id',ignore_index=False,
               var_name='transactionID',
               value_name="products")
```

```python
temp.dropna(inplace=True)
```

```python
temp.drop(columns='transactionID', axis=1, inplace=True)
```

```python
import numpy as np 
total_transactions = temp.id.nunique()
print(total_transactions)
```

```python
products_count = temp.products.nunique()
print(products_count)
```

```python
from itertools import combinations
```

```python
# Generate all pairs of products where Product1 > Product2
basket_analysis = [(p1, p2) for p1, p2 in combinations(temp['products'], 2) if p1 > p2]
```

```python
# Convert the result to a DataFrame if needed
# basket_analysis_df = pd.DataFrame(basket_analysis, columns=['Product1', 'Product2'])
```

```python
prod1 = []
prod2 = []
for i,j in basket_analysis:
    prod1.append(i)
    prod2.append(j)
```
